# Bitly-USA-Government-Website-Project
## Problem Descripition 

In 2012, URL shortening service Bitly partnered with the US government website USA.gov to provide a feed of anonymous data gathered from users who shorten links ending with .gov or .mil. 

--------------------------------------------------------------------------------------
The  file comes in JSON format, and here are some keys and their description to understand more the data and its column names. They are only the most important ones for this Project.


|key| description |
|---|-----------|
| a|Denotes information about the web browser and operating system|
| tz | time zone |
| r | URL the user come from |
| u | URL where the user headed to |
| t | Timestamp when the user start using the website in UNIX format |
| hc | Timestamp when user exit the website in UNIX format |
| cy | City from which the request intiated |
| ll | Longitude and Latitude |

# GETTING STARTED

```python
# importing Packages [Pandas, `Numpy]
import pandas as pd
import numpy as np
```

```python
# importing json package to be able to work with this json file
import json 
```

```python
# importing json file & Converting it into a dataframe
df = pd.read_json('usa.gov_click_data (1).json' , lines = True)
```

```python
# checking the dataframe
df
```

---------------------------------------------------------------------------------
## Step 0 : `Web_Browser` Column
Starting to extract the web_browser name from `a` column in `df`

```python
# creating 'web_broweser' column that contain the Splitting string output before a delimeter '/' in ['a'] column
df['web_browser'] = df['a'].apply(lambda x: x.split('/')[0])
```

```python
# checking the column
df['web_browser']
```

---------------------------------------------------------------------------------
## Step 1 : `Operating_System` Column
Starting to extract the Operating_system name from `a` column in `df`

```python
# checking the column in order to extract the operating system from each row
df['a']
```

```python
# importing re package
import re
```

```python
# iterating over each row in df['a'], then splitting each value two times to get the operating_system values into a list
operating_system_list = []
for row in df['a']:
    if row == 'GoogleMaps/RochesterNY':
        row = 'google s(Rochesterny'
    operating_system_list.append(row.split(' ')[1].split('(')[1])
operating_system_list
```

```python
# creating a column to be containing operating_system values in my df
df['operating_system'] = operating_system_list
```

```python
# cleaning the ['operating_system'] column 
df['operating_system'].replace('Macintosh;' , 'Macintosh' , inplace = True)
```

```python
# validating the new column of 'operating_system' is cleaned
df['operating_system']
```

```python
# checking the first two columns together
df[['web_browser' , 'operating_system']]
```

---------------------------------------------------------------------------------
## Step 2 : `From_URL` Column
Starting to extract the From_URL values from `r` column in `df`

```python
# checking the 'r' column original values
df['r']
```

```python
# creating an empty list to be including the 'from_url' values that the user came from
from_url = []
```

```python
# iterating over the df['r'] to extract the urls from it and appending them to from_url list
for row in df['r']:
    if row == 'direct':
        row = 'http://www.direct.com'
    from_url.append(row.split('/')[2])
# checking the list output
from_url
```

```python
# creating a column that includes the 'from_url' values in my df
df['from_url'] = from_url
df['from_url']
```

```python
# checking on the 3 columns together ['web_browser' , 'operating_system', 'from_url']
df[['web_browser' , 'operating_system' , 'from_url']]
```

---------------------------------------------------------------------------------
## STEP 3 : `To_URL` Column
Starting to extract the To_URL name from `u` column in `df`

```python
# checking the df['u'] original values to extract the shortenl url from it
df['u']
```

```python
# creating an empty list to be including the 'to_url' values
to_url = []
```

```python
# iterating over the rows in df['u'] to get the 'to_url' from it
for row in df['u']:
    to_url.append(row.split('/')[2])
# showing the list appended values    
to_url
```

```python
# saving this list into a column['to_url'] in my df
df['to_url'] =to_url
df['to_url']
```

```python
# checknig the four columns that have been extracted together 
df[['web_browser' , 'operating_system' , 'from_url' , 'to_url']]
```

---------------------------------------------------------------------------------
## STEP 4 : `City` Column 
Starting to extract the City name from `cy` column in `df`

```python
# checking on the original column before extracting data from it
df['cy']
```

```python
# so, it is clear that this column does not need any change
# assigning the df['cy'] values to another column called 'city'
df['city'] = df['cy']
df['city']
```

```python
# cehcking on all created columns together
df[['web_browser' , 'operating_system' , 'from_url' , 'to_url' , 'city']]
```

---------------------------------------------------------------------------------
## STEP 5 : `Longitude` and `Latitude` Column
Starting to extract the Longitude and Latitude from `ll` column in `df`

```python
# checking on the original column before extracting data from it
df['ll']
```

```python
# editing the indexes that have NaN values [7, 8 and 9] to be in this format '[Nan , Nan]' to be able to extract them
df['ll'][7]=  [np.nan , np.nan]
df['ll'][8]=  [np.nan , np.nan]
df['ll'][9]=  [np.nan , np.nan]
```

```python
# validing that all column's values have the same format right now
df['ll']
```

```python
# creating an empty list to be including the 'longitude' values
longitude = []
longitude
```

```python
# iterating over the rows in df['ll'] column to get only the longitude value 
for row in df['ll']:
    longitude.append(row[0])
```

```python
longitude
```

```python
# creating a column to be carrying the 'longitude' values
df['longitude'] = longitude
df['longitude']
```

```python
# creating an empty list to be containing the 'latidude' values
latitude = []
latitude
```

```python
# iterating over the rows in df['ll'] column to get only the latidude values 
for row in df['ll']:
    latitude.append(row[1])
```

```python
latitude
```

```python
# creating a column to be carrying the 'latidude' values
df['latitude'] = latitude
df['latitude']
```

```python
# checking all columns created together
df[['web_browser' , 'operating_system' ,'from_url', 'to_url', 'city' , 'longitude' , 'latitude']]
```

---------------------------------------------------------------------------------
## STEP 6:`Time Zone` Column
Starting to extract the TimeZone from `tz` column in `df`

```python
# checking the original column of 'tz' to determine which changes should happen
df['tz']
```

```python
# replacing missing values with NaN values
df['tz'][7] = np.nan
df['tz'][8] = np.nan
df['tz'][9] = np.nan
# checking the column now
df['tz']
```

```python
# creating a new column to be include the values in 'tz' column
df['time_zone'] = df['tz']
df['time_zone']
```

-------------------------------------------------------------------------------------------
## Another work-around
## Creating `time_zone_city` column based on the `city` value that user has entered, Utilizing `.map()` function
Searching for each `time-zone` for each `city` value in my df,
Then mapping them together based on the `city` column.

```python
# creating a new column ['time_zone_city'] that contain the time-zone value for each user city
df['time_zone_city'] = df['city'].map({
    'Danvers' : 'Danvers, MA, USA (GMT-5)',
    'Prove' : 'Provo, UT, USA (GMT-7)',
    'Washington' : 'Washington, DC, USA (GMT-5)',
    'Braz' : 'Brazil , Fernando de Noronha time (UTC−02:00)',
    'Shrewsbury' : 'Shrewsbury, UK (GMT)',
    'Luban' :  'Luban, Poland (GMT+1)',
    'Seattle' : 'Seattle, WA, USA (GMT-8)',
    'Washington' : 'Washington, DC, USA (GMT-5)'
}) 
# checking the column
df['time_zone_city']
```

```python
# checking the dataframe till now
df[['web_browser' , 'operating_system', 'from_url' , 'to_url', 'city' , 'longitude' , 'latitude', 'time_zone' , 'time_zone_city']]
```

---------------------------------------------------------------------------------
## STEP 7 : `Time-in` & `Time-out` Columns
Starting to extract Time-in and Time-out values from `t` and `hc` column in my `df`

```python
# checking the original column ['t'] before changing and extracting it into a new column
df['t']
```

```python
# importing datetime package to be able to convert these timestamp values into t
from datetime import datetime, timedelta
import datetime
```

```python
# iterating over the df['t'] column to convert each value to a data format
time_in = []
for row in df['t']:
    time_in.append(datetime.datetime.fromtimestamp(row))
time_in
```

```python
# creating a new column in the dataframe to be containing the values in time_in list
df['time_in'] = time_in
df['time_in']
```

```python
# iterating over the df['hc'] column to convert each value to a date format
time_out = []
for row in df['hc']:
    time_out.append(datetime.datetime.fromtimestamp(row))
time_out
```

```python
# creating a new column in the dataframe to be containing the values in time_out list
df['time_out'] = time_out
df['time_out']
```

```python
# viewing all the columns till now
df[['web_browser' , 'operating_system', 'from_url' , 'to_url', 'city', 'longitude' ,
    'latitude', 'time_zone', 'time_zone_city', 'time_in', 'time_out']]
```

---------------------------------------------------------------------------------
## STEP 8 : Importing all these cleaned columns into a new dataframe `users_gov_website` and excluding `nan` values

```python
# creating the new dataframe 
user_gov_website = df[['web_browser' , 'operating_system', 'from_url' , 'to_url', 'city', 'longitude' ,
    'latitude', 'time_zone', 'time_zone_city','time_in', 'time_out']]
```

```python
# droping nan values
user_gov_website.dropna(axis = 0 , inplace = True)
```
